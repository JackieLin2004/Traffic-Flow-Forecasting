```python
import re
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 读取txt文件
file_path = './astgcn.txt'  # 替换成你的文件路径
with open(file_path, 'r') as file:
    data = file.read()

# 2. 使用正则表达式提取 global step 和对应的训练损失
pattern = r'global step: (\d+), training loss: ([\d.]+)'  # 匹配 global step 和训练损失
matches = re.findall(pattern, data)

# 3. 提取global step和训练损失
steps = [int(match[0]) for match in matches]
losses = [float(match[1]) for match in matches]

# 4. 设置Seaborn风格背景为dark-grid
sns.set_style('darkgrid')

# 5. 绘制折线图
plt.figure(figsize=(10, 6))
plt.plot(steps, losses, marker='o', color='b', linestyle='-', markersize=6)

# 6. 设置图表标题和标签
plt.title('ASTGCN Training Loss vs Global Step')
plt.xlabel('Global Step')
plt.ylabel('Training Loss')

# 7. 保存图像到文件
output_file = './astgcn.png'  # 保存的文件路径和名称，可以修改为其他格式，如 .pdf, .svg
plt.savefig(output_file)

# 8. 显示图表
plt.show()

print(f"图像已保存为 {output_file}")
```

```python
import re
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 读取txt文件
file_path_1 = './astgcn.txt'  # 替换成第一个txt文件的路径
file_path_2 = './mstgcn.txt'  # 替换成第二个txt文件的路径

def extract_loss(file_path):
    with open(file_path, 'r') as file:
        data = file.read()
    
    # 使用正则表达式提取 global step 和对应的训练损失
    pattern = r'global step: (\d+), training loss: ([\d.]+)'  # 匹配 global step 和训练损失
    matches = re.findall(pattern, data)

    # 提取global step和训练损失
    steps = [int(match[0]) for match in matches]
    losses = [float(match[1]) for match in matches]
    
    return steps, losses

# 2. 获取两个文件的数据
steps_1, losses_1 = extract_loss(file_path_1)
steps_2, losses_2 = extract_loss(file_path_2)

# 3. 设置Seaborn风格背景为dark-grid
sns.set_style('darkgrid')

# 4. 创建一个1行2列的子图布局
fig, axes = plt.subplots(1, 2, figsize=(15, 6))  # 1行2列

# 5. 绘制第一个图
axes[0].plot(steps_1, losses_1, marker='o', color='b', linestyle='-', markersize=6)
axes[0].set_title('ASTGCN Training Loss vs Global Step')
axes[0].set_xlabel('Global Step')
axes[0].set_ylabel('Training Loss')

# 6. 绘制第二个图
axes[1].plot(steps_2, losses_2, marker='o', color='r', linestyle='-', markersize=6)
axes[1].set_title('MSTGCN Training Loss vs Global Step')
axes[1].set_xlabel('Global Step')
axes[1].set_ylabel('Training Loss')

# 7. 自动调整子图间距
plt.tight_layout()

# 8. 保存合并图像
output_file = './combined_training_loss.png'  # 保存的文件路径和名称
plt.savefig(output_file)

# 9. 显示合并的图表
plt.show()

print(f"合并图像已保存为 {output_file}")
```
