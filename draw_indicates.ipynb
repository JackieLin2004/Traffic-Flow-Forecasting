```python
import re
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'logs/astgcn.txt'
with open(file_path, 'r') as file:
    data = file.read()

pattern = r'global step: (\d+), training loss: ([\d.]+)'
matches = re.findall(pattern, data)

steps = [int(match[0]) for match in matches]
losses = [float(match[1]) for match in matches]

sns.set_style('darkgrid')

plt.figure(figsize=(10, 6))
plt.plot(steps, losses, marker='o', color='b', linestyle='-', markersize=6)

plt.title('ASTGCN Training Loss vs Global Step')
plt.xlabel('Global Step')
plt.ylabel('Training Loss')

output_file = 'logs/astgcn.png'
plt.savefig(output_file)

plt.show()

print(f"图像已保存为 {output_file}")
```

```python
import re
import matplotlib.pyplot as plt
import seaborn as sns

file_paths = {
    'PEMS04_ASTGCN': './logs/pems04_astgcn.txt',
    'PEMS04_MSTGCN': './logs/pems04_mstgcn.txt',
    'PEMS08_ASTGCN': './logs/pems08_astgcn.txt',
    'PEMS08_MSTGCN': './logs/pems08_mstgcn.txt'
}

def extract_loss(file_path):
    """从日志文件中提取训练步数和损失值"""
    with open(file_path, 'r') as file:
        data = file.read()

    pattern = r'global step: (\d+), training loss: ([\d.]+)'
    matches = re.findall(pattern, data)

    steps = [int(match[0]) for match in matches]
    losses = [float(match[1]) for match in matches]
    
    return steps, losses

steps_04_ast, losses_04_ast = extract_loss(file_paths['PEMS04_ASTGCN'])
steps_04_mst, losses_04_mst = extract_loss(file_paths['PEMS04_MSTGCN'])
steps_08_ast, losses_08_ast = extract_loss(file_paths['PEMS08_ASTGCN'])
steps_08_mst, losses_08_mst = extract_loss(file_paths['PEMS08_MSTGCN'])

sns.set_style('darkgrid')

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

axes[0, 0].plot(steps_04_ast, losses_04_ast, marker='o', color='b', linestyle='-', markersize=6)
axes[0, 0].set_title('ASTGCN Training Loss in PEMS04')
axes[0, 0].set_xlabel('Global Step')
axes[0, 0].set_ylabel('Training Loss')

axes[0, 1].plot(steps_04_mst, losses_04_mst, marker='o', color='r', linestyle='-', markersize=6)
axes[0, 1].set_title('MSTGCN Training Loss in PEMS04')
axes[0, 1].set_xlabel('Global Step')
axes[0, 1].set_ylabel('Training Loss')

axes[1, 0].plot(steps_08_ast, losses_08_ast, marker='o', color='b', linestyle='-', markersize=6)
axes[1, 0].set_title('ASTGCN Training Loss in PEMS08')
axes[1, 0].set_xlabel('Global Step')
axes[1, 0].set_ylabel('Training Loss')

axes[1, 1].plot(steps_08_mst, losses_08_mst, marker='o', color='r', linestyle='-', markersize=6)
axes[1, 1].set_title('MSTGCN Training Loss in PEMS08')
axes[1, 1].set_xlabel('Global Step')
axes[1, 1].set_ylabel('Training Loss')

plt.tight_layout()

output_file = './images/all_combined_training_loss.png'
plt.savefig(output_file)

plt.show()

print(f"合并图像已保存为 {output_file}")
```
